```python
#Collecting all the links to our chosen genres
main_genre_url = 'https://www.goodreads.com/shelf/show/'
genres_list = ['art', 'romance', 'fantasy', 'horror', 'crime', 'science', 'mystery', 'history']
genres_links = []

for genre in genres_list:
    genre_link = main_genre_url + genre
    genres_links.append(genre_link)
    
```

```python
#Generating the links to each page of books for every genre
art_page_links = []
romance_page_links = []
fantasy_page_links = []
horror_page_links = []
crime_page_links = []
science_page_links = []
mystery_page_links = []
history_page_links = []

i = 0
for genre_link in genres_links:
    if (i == 0):
        for j in range(1, 25):
            art_page_url = genres_links[i]+'?page='+ str(j) 
            art_page_links.append(art_page_url)
            
    if (i == 1):
        for j in range(1, 25):
            romance_page_url = genres_links[i]+'?page='+ str(j) 
            romance_page_links.append(romance_page_url)
    
    if (i == 2):
        for j in range(1, 25):
            fantasy_page_url = genres_links[i]+'?page='+ str(j) 
            fantasy_page_links.append(fantasy_page_url)
    
    if (i == 3):
        for j in range(1, 25):
            horror_page_url = genres_links[i]+'?page='+ str(j) 
            horror_page_links.append(horror_page_url)
    
    if (i == 4):
        for j in range(1, 25):
            crime_page_url = genres_links[i]+'?page='+ str(j) 
            crime_page_links.append(crime_page_url)
            
    if (i == 5):
        for j in range(1, 25):
            science_page_url = genres_links[i]+'?page='+ str(j) 
            science_page_links.append(science_page_url)
            
    if (i == 6):
        for j in range(1, 25):
            mystery_page_url = genres_links[i]+'?page='+ str(j) 
            mystery_page_links.append(mystery_page_url)
            
    if (i == 7):
        for j in range(1, 25):
            history_page_url = genres_links[i]+'?page='+ str(j) 
            history_page_links.append(history_page_url)
    i += 1
    
```

```python
#A function to generate the links to the actual books
def extract_info(links):
    links_to_books = []
    
    for link in links:
        page_res = requests.get(link, cookies=cookies)
        page_soup = BeautifulSoup(page_res.content, "html.parser")     
        book_containers = (page_soup.find_all("div", attrs = {"class" : "left"}))
         
        for book in book_containers:
            links_to_books.append(book.find('a').get('href'))
     
    print(len(links_to_books))
    return links_to_books
```

```python
#Function calls
links_to_art_books = extract_info(art_page_links)
links_to_romance_books = extract_info(romance_page_links)
links_to_fantasy_books = extract_info(fantasy_page_links)
links_to_horror_books = extract_info(horror_page_links)
links_to_crime_books = extract_info(crime_page_links)
links_to_science_books = extract_info(science_page_links)
links_to_mystery_books = extract_info(mystery_page_links)
links_to_history_books = extract_info(history_page_links)
```

```python
#Extracting all the information we need about each book

domain_url = "https://www.goodreads.com"

all_book_links = [links_to_art_books, links_to_romance_books, links_to_fantasy_books, links_to_horror_books, 
                 links_to_crime_books, links_to_science_books, links_to_mystery_books, links_to_history_books]

titles = []
generes = []
authors = []
num_pages = []
publish_dates = []
languages = []
num_reviews = []
num_ratings = []
ratings = []

i = 0

for gen_link in all_book_links:

    for link in gen_link:  
        print(i)
        book_link = domain_url + link
        book_res = requests.get(book_link, cookies=cookies)
        book_soup = BeautifulSoup(book_res.content, "html.parser")


        try:
            titles.append(book_soup.find('h1', attrs={"itemprop":"name"}).text.strip())
        except:
            titles.append(None)

        try:
            authors.append(book_soup.find('span', attrs={"itemprop":"name"}).text.strip()) 
        except:
            authors.append(None)

        try:
            str = book_soup.find('span', attrs={"itemprop":"numberOfPages"}).text.strip()
            num_pages.append([int(s) for s in str.split() if s.isdigit()][0])
        except:
            num_pages.append(None)

        try:
            str = book_soup.find_all('div', attrs={"class":"row"})[1].text.strip()
            publish_dates.append([int(s) for s in str.split() if s.isdigit()][0])
        except:
            publish_dates.append(None)

        try:
            languages.append(book_soup.find('div', attrs={"itemprop":"inLanguage"}).text.strip())
        except:
            languages.append(None)     

        try:
            num_reviews.append(book_soup.find('meta', attrs={"itemprop":"reviewCount"}).get('content'))
        except:
            num_reviews.append(None)    

        try:
            num_ratings.append(book_soup.find('meta', attrs={"itemprop":"ratingCount"}).get('content'))
        except:
            num_ratings.append(None)    

        try:
            ratings.append(book_soup.find('span', attrs={"itemprop":"ratingValue"}).text.strip())
        except:
            ratings.append(None)    

        time.sleep(0.5)
        i += 1

    time.sleep(90)
```

```python
#Building the DataFrame with all the information we gathered
df = pd.DataFrame(columns = ['Title','Genre', 'Author', 'Pages', 'Publish_Date', 
                                     'language', 'Num_Reviews', 'Num_Ratings', 'Rating'])
df['Title'] = titles
df['Genre'] = genres
df['Author'] = authors
df['Pages'] = num_pages
df['Publish_Date'] = publish_dates
df['language'] = languages
df['Num_Reviews'] = num_reviews
df['Num_Ratings'] = num_ratings
df['Rating'] = ratings
```
